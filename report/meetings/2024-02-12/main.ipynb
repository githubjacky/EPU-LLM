```python
import pandas as pd


category = 'policy'
countries = ['TW', 'CN', 'JP', 'KR', 'US', 'AU']
model = 'chatgpt3.5'

col1 = list(compare_human(category, 'vote', countries, model).values())
col2 = list(compare_human(category, 'definition', countries, model).values())

N = max(len(countries), len(col1), len(col2))
df = pd.DataFrame({
    'countries': countries + ['']*(N-len(countries)),
    'vote': col1 + ['']*(N-len(col1)),
    'definition': col2 + ['']*(N-len(col2))
})

print(df.to_latex(index=False))
```

```python
col3 = list(compare_human(category, 'definition with SC', countries, model).values())
N = max(len(countries), len(col1), len(col2), len(col3))
df = pd.DataFrame({
    'countries': countries + ['']*(N-len(countries)),
    'vote': col1 + ['']*(N-len(col1)),
    'definition': col2 + ['']*(N-len(col2)),
    'definition with SC': col3 + ['']*(N-len(col3))
})

print(df.to_latex(index=False))
```
